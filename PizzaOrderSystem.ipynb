```python
# pizza üst sınıfı
class pizza:
  def __init__(self, description, cost):
     self.__descripton = description
     self.__cost  = cost
  def get_description(self):
    return self.__descripton
  def get_cost(self):
    return self.__cost
```

```python
# pizza çeşitleri alt sınıfları
class KlasikPizza(pizza):
    def __init__(self):
        description = "Kalın hamur, ince kenar, sosis, kaşar, zeytin, biber, mısır"
        cost = 98.95
        super().__init__(description, cost)

class TurkishPizza(pizza):
    def __init__(self):
        description = "İnce hamur, kalın kenar, sucuk, kaşar, biber, mantar, mısır, sosis"
        cost = 129.90
        super().__init__(description, cost)

class MargaritaPizza(pizza):
    def __init__(self):
        description = "İnce hamur, ince kenar, domates, mozarella peyniri, fesleğen"
        cost = 134.98
        super().__init__(description, cost) 

class SadePizza(pizza):
    def __init__(self):
        description = "Kalın hamur, kalın kenar, kaşar"
        cost = 59.99
        super().__init__(description, cost)

class TonBalikliPizza(pizza):
    def __init__(self):
        description = "İnce hamur, ince kenar, kaşar, ton balığı, mısır"
        cost = 130.44
        super().__init__(description, cost)
```

```python
#ekstra sos sınıfları
class Decorator(pizza):
    def __init__(self, pizza):
        self._pizza = pizza

    def get_description(self):
        return self._pizza.get_description()

    def get_cost(self):
        return self._pizza.get_cost()


class Zeytin(Decorator):
    def __init__(self, pizza):
        cost = 10
        description = "Zeytin"
        super().__init__(pizza)
        self._description = pizza.get_description() + ", " + description
        self._cost = pizza.get_cost() + cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost


class Mantar(Decorator):
    def __init__(self, pizza):
        cost = 15
        description = "Mantar"
        super().__init__(pizza)
        self._description = pizza.get_description() + ", " + description
        self._cost = pizza.get_cost() + cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost


class KeciPeyniri(Decorator):
    def __init__(self, pizza):
        cost = 25
        description = "Keçi Peyniri"
        super().__init__(pizza)
        self._description = pizza.get_description() + ", " + description
        self._cost = pizza.get_cost() + cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost


class Et(Decorator):
    def __init__(self, pizza):
        cost = 30
        description = "Et"
        super().__init__(pizza)
        self._description = pizza.get_description() + ", " + description
        self._cost = pizza.get_cost() + cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost

class Sogan(Decorator):
    def __init__(self, pizza):
        cost = 8
        description = "Soğan"
        super().__init__(pizza)
        self._description = pizza.get_description() + ", " + description
        self._cost = pizza.get_cost() + cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost


class Misir(Decorator):
    def __init__(self, pizza):
        cost = 12
        description = "Mısır"
        super().__init__(pizza)
        self._description = pizza.get_description() + ", " + description
        self._cost = pizza.get_cost() + cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost
```

```python
def main():
    with open("Menu.txt", "r") as menu:
        print(menu)


if __name__ == '__main__':
    pizza = None

    while not pizza:
        try:
            base_choice = int(input(
                "Lütfen bir pizza tabanı seçiniz: (1-Klasik Pizza, 2-Turkish Pizza, 3-Margarita Pizza, 4-Sade Pizza, 5-Ton Balıklı Pizza) "))
            if base_choice == 1:
                pizza = KlasikPizza()
            elif base_choice == 2:
                pizza = TurkishPizza()
            elif base_choice == 3:
                pizza = MargaritaPizza()
            elif base_choice == 4:
                pizza = SadePizza()
            elif base_choice == 5:
                pizza = TonBalikliPizza()
            else:
                print("Lütfen geçerli bir seçim yapınız.")
        except ValueError:
            print("Lütfen geçerli bir sayı giriniz.")

    while True:
        try:
            sos_choice = int(input(
                "Lütfen bir sos seçiniz: (11-Zeytin, 12-Mantarlar, 13-Keçi Peyniri, 14-Et, 15-Soğan, 16-Mısır) "))
            if sos_choice == 11:
                pizza = Zeytin(pizza)
            elif sos_choice == 12:
                pizza = Mantar(pizza)
            elif sos_choice == 13:
                pizza = KeciPeyniri(pizza)
            elif sos_choice == 14:
                pizza = Et(pizza)
            elif sos_choice == 15:
                pizza = Sogan(pizza)
            elif sos_choice == 16:
                pizza = Misir(pizza)
            else:
                print("Lütfen geçerli bir seçim yapınız.")
            break
        except ValueError:
            print("Lütfen geçerli bir sayı giriniz.")

    aciklama = input("Açıklama: ")
      # Sipariş zamanını
    order_time = datetime.datetime.now()
    
    print("------------------------")
    print("Siparişiniz hazır!")
    print("Sipariş tarihi:", order_time.strftime("%d-%m-%Y %H:%M"))
    print("Sipariş detayı:", pizza.get_description())
    print("Toplam fiyat:", pizza.get_cost(), "TL")
    print("Açıklama: ", aciklama)
    print("------------------------")
    print()
    # Kullanıcı bilgileri alınır
    print("------ÖDEME ALANI------")
    name = input("İsim-Soyisim: ")
    tc_no = input("TC Kimlik Numaranız: ")
    while len(tc_no) != 11:
        print("Geçersiz TC Kimlik Numarası!")
        tc_no = input("TC Kimlik Numarası: ")
    cc_no = input("Kredi Kart Numarası: ")
    cc_password = input("Kredi Kart Şifresi: ")
    while len(cc_password) != 4:
        print("Geçersiz kredi kartı şifresi! Lütfen 4 haneli şifre giriniz!")
        cc_password = input("Kredi Kartı Şifresi: ")
    print("------------------------")
    print()
    print("Merhaba " + name + " siparişiniz oluşturulmuştur. ")
    print()
```
